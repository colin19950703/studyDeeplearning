##[] OOP (Object Oriented Programming) </br></br>

OOP (Object Oriented Programming)란?
객체 지향 프로그래밍이라는 뜻으로 실체가 있는 모든 것을 클래스와 인스턴스. 함수와 변수라는 오브젝트로 변화 시켜서 프로그래밍을 한다고 생각하면 될것이다.</br>
OOP의 목적은 재사용성에 있기 때문에 프로그램이 분리가 되도록 설계를 해야한다.



###[]클래스 <br><br>

클래스는 쉽게 말해 조립 설명서와 같다. <br>
하나의 제품을 만든다고 가정했을 때, 어떤 부품이 어떻게 연결되는지를 설명해놓은 조립 설명서와 같다고 보면 된다. <br>
반면 객체화는 이 조립 설명서대로 제품을 찍어내는 것을 말한다. <br><br>

 이것을 컴퓨터의 입장에서 보면 클래스는 메모리상에 존재하지 않지만 객체(제품)이 어떤 속성과 멤버 함수를 가져야하는지 설명 해놓은 것이다.<br>
 그리고 객체화는 이 클래스가 컴퓨터의 메모리 상에 공간을 차지하게 되는 순간을 말한다.<br>
 이러한 과정을 거치게 되면 번거롭지만, 이 과정을 통해 클래스에 의한 객체의 재사용성과 범용성이 증가된다.<br><br>
 즉, 잘 만들어 놓은 클래스(조립 설명서)하나로 같은 형태의 여러 객체(제품)을 찍어 낼 수 있다.<br>
 게다가 상속과 오버 로딩을 통해서 조금 변형이 된 새로운 제품도 쉽게 만들어 낼수 있게 된다.
 

###[]캡슐화<br>
<img src = "https://github.com/Maiven/data-science/blob/main/%E1%84%8F%E1%85%A2%E1%86%B8%E1%84%89%E1%85%B2%E1%86%AF%E1%84%92%E1%85%AA.png?raw=true" width = '400'><br>
비슷한 역할을 하는 속성과 메소드들을 하나의 클래스로 모은것을 캡슐화 라고 한다.
<br><br>
내부 속성(변수)과 함수를 하나로 묶어서 클래스로 선언하는 개념. <br>
(하나의 카테고리안에서 하위 카테고리의 기능들을 하나로 묶어 클래스로 선언)<br>
캡슐화형태로 코드를 작성하지 않으면 특정 기능(함수, 변수)에 직접 접근하게 되는 상황이 된다.<br>
기능이 많아질수록 재사용의 개념을 활용하기가 어려움<br>



###[]상속과 포함<br>
<img src = "https://github.com/Maiven/data-science/blob/main/inheriance_composition.png?raw=true" width = '400'><br>
상속이란 클래스를 재사용 하는것이다. 상위 클래스를 하위 클래스에서 상속 받게 되면 상위 클래스의 멤버변수나 메소드를 그대로 물려 받을 수 있다.<br>
상속이 있기 때문에 코드를 재활용할 수 있고 그렇기 때문에 생산성이 높고 유지보수 하기가 좋다. <br><br>
상속 : 기존에 정의 되어있는 클래스의 모든 멤버변수와 멤버함수를 물려받아, 새로운 클래스를 작성하는 것이다. <br>

포함 : 다른 클래스의 일부 기능(함수)만을 재사용한다.<br><br>

기존의 정의된 클래스 : 부모 클래스, 기초 클래스, 상위 클래스<br>
상속을 통해 새롭게 작성된 클래스 : 자식 클래스, 파생 클래스, 하위 클래스<br>




###[]추상화<br>
<img src = "https://github.com/Maiven/data-science/blob/main/%E1%84%8E%E1%85%AE%E1%84%89%E1%85%A1%E1%86%BC%E1%84%92%E1%85%AA.png?raw=true" width = '400'><br>
복잡한 내용에서 핵심적인 개념 및 기능을 요약하는 것을 말한다.<br>
예를들어서, 지구를 본따 만든 지구본을 예로 들 수 있다. 지구본은 실제 지구로 부터 관심 있는 특성들(대륙의 위치, 위도,경도)만 뽑아서 만든것이다.<br>
지구를 추상화해서 지구본을 만들었다.<br>
객체지향에서의 추상화는 어떤 하위클래스들에 존재하는 공통적인 메소드를 인터페이스로 정의하는것을 예로 들 수 있다.<br><br>


object의 기능에 따라 추상클래스(부모클래스)를 상속받아 개별적으로 클래스(자식클래스)를 생성한다.<br>
기본적으로 추상메소드를 선언하며 실제 실행되는 기능은 보여지지 않는다.<br>
실제 실행되는 기능은 선언된 추상클래스를 상속받은 다른 클래스의 메소드에서 확인할 수 있다.<br>
[추상클래스를 사용하는 이유]<br>
대형 프로젝트를 진행하는 경우 또는 프로그램이 복잡해지는 경우 1차적인 설계를 위해 기능을 추상화시켜놓고, 활용여부는 차후 결정하기 위함이다.<br>




###[]다형성<br>
<img src = "https://github.com/Maiven/data-science/blob/main/%E1%84%83%E1%85%A1%E1%84%92%E1%85%A7%E1%86%BC%E1%84%89%E1%85%A5%E1%86%BC.png?raw=true" width = '400'><br>
구현되는 하위클래스에 따라 클래스를 다르게 처리하는 기능<br>
상속과 유사하다고 느껴질 수 있지만, 상속은 상위클래스의 기능(함수, 변수)을 재사용한다.<br>
위의 그림과 같이 다형성은 상위클래스의 기능을 변경하여 사용하는 것이다.(그대로 재사용하지 않는다.)<br>

다형성은, 같은 모양의 함수가 상황에 따라 다르게 동작 하는것을 의미한다. 오버로딩과 오버라이딩이 있다.<br><br>

오버로딩 : 서로 다른 시그니처를 갖는 여러 개 함수를 같은 이름으로 정의하는 것.<br>
[조건]<br>
1. 함수명은 동일해야 한다. <br>
2.매개변수는 상이해야 한다. <br>
3.리턴타입은 상관없다.<br>

오버라이딩 : 이미 정의된 함수를 무시하고 같은 이름의 함수를 새롭게 정의하는 것.<br>
항수의 동작만을 재정의 하기 때문에 함수 원형은 기존 멤버함수의 원형과 같아야 한다.<br>
